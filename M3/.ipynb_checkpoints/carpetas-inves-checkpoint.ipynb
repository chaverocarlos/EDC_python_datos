```python
delitos_robo = data[data.delito.str.contains("ROBO")]
```

```python
delitos_robo
```

```python
delitos_robo_r = delitos_robo[["delito", "fecha_hechos", "fiscalia"]]
```

```python
delitos_robo_r["fecha_hechos"] = pd.to_datetime(delitos_robo_r["fecha_hechos"], format="%Y-%m-%d %H:%M:%S")
```

```python
# delitos_robo_r.groupby("delito").count().nsmallest(10, 'fiscalia')
array_del = delitos_robo_r.groupby("delito").count().nlargest(10, 'fiscalia').index
```

```python
for i in array_del:
    delito_tmp = delitos_robo_r[delitos_robo_r["delito"] == i]
    x = delito_tmp.groupby(delito_tmp["fecha_hechos"].dt.hour).count().index
    y = delito_tmp.groupby(delito_tmp["fecha_hechos"].dt.hour).count()["fiscalia"]
    plt.figure()
    plt.title(i)
    plt.stem(x, y)
    nombre = 'robos/'+i+'.jpg'
    plt.savefig(nombre, bbox_inches="tight")
```
