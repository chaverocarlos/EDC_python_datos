```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("carpetas-investigacion.csv")
```

```python
data
```

```python
data["alcaldia_hechos"].unique()
```

```python
#delitos de Azcapotzalco
alcIzt = data[data['alcaldia_hechos'] == 'IZTAPALAPA']
alcIzt
```

```python
alcIzt["mes_hechos"].unique()
```

```python
alcIzt = alcIzt[alcIzt["mes_hechos"].notna()]
```

```python
alcIzt["mes_hechos"].unique()
```

```python
resumen = alcIzt[['delito', 'colonia_hechos', 'fecha_hechos']]
```

```python
resumen.groupby("colonia_hechos").count().nlargest(10, 'delito')
```

```python
resumen["fecha_hechos"] = pd.to_datetime(resumen["fecha_hechos"], format="%Y-%m-%d %H:%M:%S")
```

```python
resumen.groupby( resumen["fecha_hechos"].dt.month ).count().nlargest(12, 'delito')
```

```python
x = resumen.groupby( resumen["fecha_hechos"].dt.hour ).count().index
y = resumen.groupby( resumen["fecha_hechos"].dt.hour ).count()["delito"]
```

```python
plt.stem(x,y)
```

```python
x = resumen.groupby( resumen["fecha_hechos"].dt.month ).count().index
y = resumen.groupby( resumen["fecha_hechos"].dt.month ).count()["delito"]
```

```python
plt.bar(x,y)
```
